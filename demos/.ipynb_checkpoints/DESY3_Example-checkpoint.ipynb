```python
from CombineHarvester import Harvest
import numpy as np
from matplotlib import pyplot as plt
import getdist.plots as gdplt
from getdist import MCSamples
```

```python
data = np.genfromtxt('../demos_data/chain_3x2pt_lcdm_SR_maglim.txt',names=True)
```

```python
cosmo_chain = np.zeros((data.shape[0],4))
weights = data['weight']
cosmo_chain[:,0] = data['cosmological_parametersomega_m'] 
cosmo_chain[:,1] = data['cosmological_parametersa_s'] 
cosmo_chain[:,2] = data['cosmological_parametersh0']
cosmo_chain[:,3] = data['cosmological_parametersomega_b'] 
```

```python
DESY3Harvest = Harvest(cosmo_chain, weights = weights, n_components=600, n_models=1)
```

```python
DESY3Harvest.train()
```
